```python
import pandas as pd
df_crypto = pd.read_csv("Top 100 Cryptocurrency 2022.csv")

for index, row in df_crypto.iterrows():
    row['Price'] = float(row['Price'].replace('$','').replace(',','').replace('"',''))
    row['Changes 24H'] = float(row['Changes 24H'].replace('%','').replace(',','').replace('"','').replace('$',''))
    df_crypto.iloc[index] = row


def delete_df_crypto_with_less_value(bitcoin_value):
    try:
        if(type(bitcoin_value) not in [int, float]):
            raise ValueError("The value must be an integer.")
        else:
            df_crypto_filtered = df_crypto.loc[(df_crypto['Price'] >= bitcoin_value)]
    except ValueError as err:
        print(err)
    return df_crypto_filtered

df_crypto.head()
```
