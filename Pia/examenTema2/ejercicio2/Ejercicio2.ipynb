```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

misscalculatedTeams = [] #Lista para guardar los equipos con los puntos mal calculados
fit_transformer = MinMaxScaler()

dfNba = pd.read_csv("nba.csv")

#filtrar dataframe por 2 criterios
dfNbaFiltered = dfNba.loc[dfNba["G"] > 45].loc[dfNba["3P"]>13]

print("Equipos de la NBA con mas de 45 partidos jugador y 13 triples por partido \n")
print(dfNba.head(10))

#Mostrar a gente que tenga mal calculados los puntos por partido
for index, row in dfNba.iterrows():
    if(row["PTS"] != (row["3P"] * 3 + 2 * row["2P"] + row["FT"])):
        misscalculatedTeams.append(row["Team"])

print("Los equipos con los puntos mal calculados son un total de {} equipos: \n {} ".format(len(misscalculatedTeams),misscalculatedTeams))

#Quitar columnas en un nuevo dataframe
dfNbaWithDroppedColumns = dfNba
dfNbaWithDroppedColumns.drop(columns=["Rk","Team","PTS"], inplace=True,axis=1)
print("DataFrame con las columnas borradas")
print(dfNbaWithDroppedColumns.head())

#Normalizar datos
dfNbaWithDroppedColumns["2P"] = fit_transformer.fit_transform(dfNbaWithDroppedColumns["2P"].to_numpy().reshape(-1,1))

#Ordenar para mostrar los 5 con mayor correlacion
dfNbaWithDroppedColumns.sort_values(by='2P', inplace=True, ascending=False)

#Mostrar los 5 primeros
print("Los 5 con mayor correlación de 2P \n")
print(dfNbaWithDroppedColumns.head())

print(dfNbaWithDroppedColumns.corr(numeric_only=True)['2P'].abs().sort_values(ascending=False)[1:6])




```
