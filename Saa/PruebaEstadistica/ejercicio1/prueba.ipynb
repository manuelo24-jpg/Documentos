```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
datos = [5, 7, 8, 8, 9, 10, 10, 14, 14, 15, 20]
print("Media:", np.mean(datos))
print("Mediana:", np.median(datos))
print("Varianza:", np.var(datos))
print("Cuartiles:", np.percentile(datos,[25, 50, 75]))
plt.boxplot(datos)
plt.show()
```

```python
# numpy, pandas and matplotlib are already imported in another cell

# load data (omit this if you want to use the existing df_items variable)
df_items = pd.read_csv("prueba.csv", sep=";")

# Clean 'Año' column:
# - replace "N.V." with empty string
# - remove any non-digit characters
# - convert empty results to NaN
df_items['Año'] = df_items['Año'].astype(str).replace('N.V.', '', regex=False)
df_items['Año'] = df_items['Año'].str.replace(r'\D+', '', regex=True)
df_items['Año'].replace('', np.nan, inplace=True)

# Convert to numeric (coerce invalid to NaN) and keep in a new column
df_items['Año_num'] = pd.to_numeric(df_items['Año'], errors='coerce')

print(df_items['Año'])

# Prepare values for statistics (drop NaNs)
vals = df_items['Año_num'].dropna().astype(float)

if vals.size == 0:
    print("No numeric 'Año' values found.")
else:
    print("Media:", np.mean(vals))
    print("Mediana:", np.median(vals))
    print("Varianza:", np.var(vals))
    print("Cuartiles:", np.percentile(vals, [25, 50, 75]))
    plt.boxplot(vals)
    plt.show()
```
