```python
"""1. Carga y exploración de datos
Importa las librerías necesarias (pandas, numpy, matplotlib, sklearn).
Carga el fichero empleados.csv.
Muestra las primeras filas del conjunto de datos.
Comprueba si existen valores nulos o inconsistencias.
"""

import pandas as pd

dfEmployees = pd.read_csv("empleados_v2.0.csv")

dfEmployees.sort_values(by=['salario'], ascending=True, inplace=True)

dfEmployees.head(20)

```

```python
"""
2. Análisis exploratorio
Calcula estadísticas descriptivas básicas.
Representa gráficamente la relación entre las variables independientes y el salario.
Comenta brevemente qué variables parecen tener mayor relación con el salario.
"""
import numpy as np
import matplotlib.pyplot as plt
import statistics

print("La media de salario de los empleados es {}".format(np.mean(dfEmployees['salario'])))
print("La media de edad de los empleados es  {}".format(np.mean(dfEmployees['edad'])))
print("La media de experiencia de los empleados es  {}".format(np.mean(dfEmployees['experiencia'])))
print("La media de formación de los empleados es  {}".format(np.mean(dfEmployees['formacion'])))

print("La mediana de salario de los empleados es {}".format(np.median(dfEmployees['salario'])))
print("La mediana de edad de los empleados es  {}".format(np.median(dfEmployees['edad'])))
print("La mediana de experiencia de los empleados es  {}".format(np.median(dfEmployees['experiencia'])))
print("La mediana de formación de los empleados es  {}".format(np.median(dfEmployees['formacion'])))

print("La moda de salario de los empleados es {}".format(statistics.mode(dfEmployees['salario'])))
print("La moda de edad de los empleados es  {}".format(statistics.mode(dfEmployees['edad'])))
print("La moda de experiencia de los empleados es  {}".format(statistics.mode(dfEmployees['experiencia'])))
print("La moda de formación de los empleados es  {}".format(statistics.mode(dfEmployees['formacion'])))

print("La varianza de salario de los empleados es {}".format(statistics.variance(dfEmployees['salario'])))
print("La varianza de edad de los empleados es  {}".format(statistics.variance(dfEmployees['edad'])))
print("La varianza de experiencia de los empleados es  {}".format(statistics.variance(dfEmployees['experiencia'])))
print("La varianza de formación de los empleados es  {}".format(statistics.variance(dfEmployees['formacion'])))

print("La desviación estándar de salario de los empleados es {}".format(statistics.stdev(dfEmployees['salario'])))
print("La desviación estándar de edad de los empleados es  {}".format(statistics.stdev(dfEmployees['edad'])))
print("La desviación estándar de experiencia de los empleados es  {}".format(statistics.stdev(dfEmployees['experiencia'])))
print("La desviación estándar de formación de los empleados es  {}".format(statistics.stdev(dfEmployees['formacion'])))

print(dfEmployees.corr(numeric_only=True)['salario'].abs().sort_values())

#Las variables que parecen tener mayor relación con el salario son la edad, la experiencia y la formación respectivamente.



```

```python
"""3. Preparación de los datos
Separa las variables independientes (X) y la variable dependiente (y). y:salario X: edad, experiencia, formacion, horas_extra
Divide el conjunto de datos en conjunto de entrenamiento y de prueba (por ejemplo, 80 % / 20 %).
Justifica brevemente por qué es necesaria esta división."""

import matplotlib.pyplot as plt

plt.scatter(dfEmployees['experiencia'],dfEmployees['salario'])
plt.title("Grafico de salario por años de experiencia")
plt.xlabel("Experiencia en años")
plt.ylabel("Salario")
plt.show()

```

```python
"""3. Preparación de los datos
Separa las variables independientes (X) y la variable dependiente (y).
Divide el conjunto de datos en conjunto de entrenamiento y de prueba (por ejemplo, 80 % / 20 %).
Justifica brevemente por qué es necesaria esta división."""


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

x = dfEmployees[['edad','experiencia','formacion','horas_extra']]
y = dfEmployees['salario']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state = 42
)
```

```python
"""4. Construcción del modelo
Crea un modelo de regresión lineal.
Entrena el modelo con los datos de entrenamiento.
Muestra la ecuación del modelo obtenido."""

model = LinearRegression()
model.fit(x_train, y_train)

pd.DataFrame({
    'Variable' : x.columns,
    'Coeficiente' : model.coef_
})
```

```python
"""5. Evaluación del modelo
Realiza predicciones sobre el conjunto de prueba.
Calcula al menos una métrica de evaluación (R2, MAE o MSE).
Interpreta el resultado obtenido."""

y_pred = model.predict(x_test)

print("R2:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```
